```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
tips.sample(10)
```

```python
iris.sample(10)
```

# Categorical Plots




###  Categorical Scatter Plot

#### StripPlot

```python
# axes level function
sns.stripplot(tips, x ='day', y='total_bill')
```

```python
sns.stripplot(tips, x ='day', y='total_bill', jitter=False)
```

```python
# figure level function
sns.catplot(tips, x ='day', y='total_bill', kind='strip', jitter=False)
```

```python
sns.catplot(tips, x ='day', y='total_bill', kind='strip', jitter=0.2, hue='sex')
```

#### swarmplot

```python
sns.catplot(tips, x ='day', y='total_bill', kind='swarm')
```

###  Categorical Distribution Plot

#### BoxPlots

```python
sns.boxplot(tips, x='day', y='total_bill')
```

```python
sns.boxplot(tips, x='day', y='total_bill', hue='sex')
```

```python
# single boxplot

sns.boxplot(tips, y='total_bill')
```

#### Violin Plot = (box plot + kde plot)

```python
sns.violinplot(tips, x='day', y='total_bill')
```

```python
sns.violinplot(tips, x='day', y='total_bill', hue='sex')
```

```python
sns.violinplot(tips, x='day', y='total_bill', hue='sex', split=True)
```

###  Categorical Estimate Plot

#### BarPlot

```python
sns.barplot(tips, x='sex', y='total_bill')
```

```python
sns.barplot(tips, x='sex', y='total_bill', hue='smoker')
```

```python

sns.barplot(tips, x='sex', y='total_bill', hue='smoker', estimator=np.sum, errorbar=None)
```

#### PointPlot

```python
sns.pointplot(tips, x='sex', y='total_bill', hue='smoker', estimator=np.min)
```

#### Countplot

```python
sns.countplot(tips, x='sex', hue='day')
```

## Regression Plot

### regplot

```python
# hue is not available
sns.regplot(tips, x='total_bill', y='tip')
```

### lmplot

```python
sns.lmplot(tips, x='total_bill', y='tip', hue='sex')
```

## Residplot

```python
sns.residplot(tips, x='total_bill', y='tip')
```

## A second way to plot faced plots

### FacetGrid

```python

g = sns.FacetGrid(tips, col='day', row='time')
g.map(sns.boxplot, 'sex', 'total_bill', order=['Male', 'Female'])
```

## Pairplot

```python
sns.pairplot(iris, hue='species')
```

## PairGrid


```python
g = sns.PairGrid(iris)
g.map(sns.scatterplot)
```

```python
# map_diag map_offdiag
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
# map upper map lower
g = sns.PairGrid(iris, hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_offdiag(sns.scatterplot)
g.map_lower(sns.scatterplot)
```

## JointPlot

```python
sns.jointplot(tips, x='total_bill', y='tip', kind='hist', hue='sex')
```

## JoindGrid

```python
g = sns.JointGrid(tips, x='total_bill', y='tip')
g.plot(sns.scatterplot, sns.histplot)
```
