# Seaborn Complete

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## Relational Plot

### ScatterPlot

```python
df1 = sns.load_dataset('tips')
df1.head(5)
```

```python
# axes level function 
plt.figure(figsize=(12,7))
sns.scatterplot(df1, x='total_bill', y = 'tip', hue='sex', style='time', size='size', sizes=(20, 200))
```

```python
# figure level function square shape   figure size
# relplot
sns.relplot(df1, x='total_bill', y = 'tip', kind='scatter', hue='sex', style='time', size='size', height=7, aspect=1.2)
```

### LinePlot

```python
df2 = px.data.gapminder()
df2.sample(5)
```

```python
temp = df2[df2['country'] == 'Pakistan']
temp
```

```python
# axes level function
sns.lineplot(temp, x='year', y='lifeExp')
```

```python
# figure level function

sns.relplot(temp,  x='year', y='lifeExp', kind='line')
```

```python
# hue style

temp = df2[df2['country'].isin(['Pakistan', 'China','India'])]
temp.sample(6)
```

```python
temp = df2[df2['country'].isin(['Pakistan', 'Brazil','Germany'])]

sns.relplot(temp, x='year', y='lifeExp', kind='line', hue='country', style='continent', size='continent')
```

## Facet Plot

```python
df1.head(3)
```

```python
# faced plot only work on figure level function 
sns.relplot(df1, x='total_bill', y = 'tip', kind='scatter', col='smoker', row='sex')
```

## col wrap

```python
sns.relplot(df2, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=3)
```

## Distribution Plots

### histplot

```python
# figure level -> displot
# axex level -> histplot kdeplot rugplot
```

```python
# axes level function
sns.histplot(df1, x='total_bill')
```

```python
# figure level function
sns.displot(df1, x='total_bill', kind='hist')
```

```python
# on categorical column

sns.displot(df1, x='day', kind='hist')
```

```python
df3 = sns.load_dataset('titanic')
df3
```

```python
sns.displot(df3, x='age', kind='hist', hue='sex')
```

```python
sns.displot(df3, x='age', kind='hist', col='sex', hue='sex')
```

## kde plot

```python
# axes level
sns.kdeplot(df1, x='total_bill')
# figure levek

sns.displot(df1, x='total_bill', kind='kde', hue='sex', fill=True)
```

```python
# bivariate histplot

sns.histplot(df1, x='total_bill', y='tip')
```

```python
# bivariate kdeplot

sns.kdeplot(df1, x='total_bill', y='tip')
```

## Matrix Plots

### HeatMap

```python
df2.sample(10)
```

```python
temp = df2.pivot_table(index='country', columns='year', values='lifeExp')
temp
```

```python
# axes level function no figure level
plt.figure(figsize=(12,8))
sns.heatmap(temp)
```

```python
# annot
temp = df2[df2['continent'] == 'Europe'].pivot_table(index='country', columns='year', values='lifeExp')
plt.figure(figsize=(16,12))
sns.heatmap(temp, annot=True, linewidth=0.5, cmap='viridis', annot_kws={"size": 12})
```
